```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("survey_results_public.csv")
df
```

```python
# Remove Null Values
df.dropna(inplace=True)
```

```python
arr=df.loc[df['Country']=='India']
plot=arr['Age']
plot
```

```python
# Plotting Histogram
```

```python
ar=(10,20,30,40,50,60,70)
plt.xlabel('Age')
plt.ylabel('Total')
plt.title('Histogram of Ages')
plt.hist(plot,bins=ar,edgecolor='black')
```

```python
val=df['Gender'].value_counts()
res=val.head(2)
label=['Male','Female']
plt.pie(res,labels=label,autopct="%.1f%%")
plt.legend(label)
plt.axis('equal')
plt.show()
```
