```python
#Which  is most used  programming langauge (use LanguageWorkedWith column)

count=df["LanguageWorkedWith"].value_counts() 
count.head(1)
```

```python
#Display how many work remotely on pie chart  [use WorkRemote column]

import matplotlib.pyplot as plt
data=df['WorkRemote'].value_counts()
plt.pie(data,labels=data,autopct='%.1f%%')
plt.legend(data,loc="upper right")
plt.axis('equal')
plt.show()
```

```python
#Display how many male and female programmers on pie chart  [use Gender column]

val=df['Gender'].value_counts()
res=val.head(2)
label=['Male','Female']
plt.pie(res,labels=label,autopct="%.1f%%")
plt.legend(label)
plt.axis('equal')
plt.show()
```

```python
#which developer type is in demand [use DevType column]

result=df['DevType'].value_counts()
print('In demand developers:',)
result.head(1)
```
