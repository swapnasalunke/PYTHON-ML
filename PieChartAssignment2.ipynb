```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd
```

```python
df=pd.read_csv("computer_parts.csv")
df
```

```python
Data=df.groupby(df['Product Name;'])['Number Of Bugs'].sum()
Label=Data.index
```

```python
plt.pie(Data,labels=Label,autopct='%.2f%%')
plt.axis('equal')
plt.title('Computer Parts')
plt.show()
```
